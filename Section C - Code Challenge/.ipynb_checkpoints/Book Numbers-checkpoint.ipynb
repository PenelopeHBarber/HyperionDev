```python
def check13(number):
    """
    Calculates the check digit for a 13-digit ISBN number.

    Args:
        number (str): The 13 digit ISBN number.

    Returns:
        result: The calculated check digit.
    """
    result = 0
    for i in range(0,13):
        if i%2==0:
            result += int(number[i])*1
        else:
            result += int(number[i])*3
    return(result)


def check10(number):
    """
    Calculates the check digit for a 10-digit ISBN number.

    Args:
        number (str): The 10 digit ISBN number.

    Returns:
        result: The calculated check digit.
    """
    result = 0
    for i in range(0,10):
        #CHECK IF THE LAST DIGIT IS AN X AND CHANGE THE VALUE TO 10
        if str.lower(number[i])=="x":
           result+= 10*(10-i)
        else:
            result += int(number[i])*(10-i)
    return(result)


def convert13(number):
    """
    Converts a 10-digit ISBN number to a 13-digit ISBN number by adding the prefix "978".

    Args:
        number (str): A 10-digit ISBN number.

    Returns:
        convert: The 13-digit ISBN number.
    """
    convert = "978"+ number
    return(convert)


def isbn13(number):
    """
    Validates and converts a given ISBN number to its 13-digit format.

    Args:
        number (str): The ISBN number to be validated and/or converted.

    Returns:
        str: "Valid" if the ISBN number is valid, otherwise the corrected 13-digit ISBN number if the given number is invalid.
    """
    if len(number)>10:
        result13 = check13(number)
        
        #check if result is divisible by 10
        if result13 %10 == 0:
            return("Valid")
        else:
            return("Invalid")
    else:
        #CHECK IF RESULT IS DIVISIBLE BY 11
        result10 = check10(number)
        #if check 10 is valid, convert to 13 and return value
        if result10 %11 == 0:
            con = convert13(number)
            new_result = check13(con)
            remainder = new_result%10
            last_digit = int(con[-1])
            if last_digit > remainder:
                new_last = last_digit-remainder
                new13 = con[:-1] + str(new_last)
                return(new13)
            else:
                new_last = last_digit+(10-remainder)
                new13 = con[:-1] + str(new_last)
                return(new13)
        
        else:
            return("Invalid")

print(isbn13("9780316066525"))
print(isbn13("9780316066526"))
print(isbn13("0330301624")) 
```
