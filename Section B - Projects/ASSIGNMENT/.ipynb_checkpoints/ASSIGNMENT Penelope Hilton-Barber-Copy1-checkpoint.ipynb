```python
#from inspection there are more than 3 columns with numerical data
#I found that the presence of "---" was causing pandas to view columns with numeric data as objects
#when I converted everything to string to replace the "---" with ""
#then the NaN values, especially in the last two columns, were a problem
```

```python
#LIST OF COLUMNS
cols = hfa.columns.to_list()
```

```python
last_two_cols = cols[-2:]
last_two_cols
```

```python
#fill in missing values in last 2 columns 
hfa.loc[:,last_two_cols] = hfa.loc[:,last_two_cols].ffill(axis="rows")

#filling in rows with values is not ideal. I would prefer to use how the form_score_total 
#and form.score_max_total were calculates and impute the sum. If possible I would contact the 
#international partner to find out moare about the dataset and how the values are calculated.
```

```python
#create new column for percentage score which will be the target
hfa["form.score_percent"] = hfa["form.score_total"] / hfa["form.score_max_total"] *100
```

```python
#check for null values
#hfa.isnull().sum()
```

```python
#remove any trailing spaces
#convert column to string type, remove spaces, then convert back to numeric
#allowing errors to be ignored lets the columns with text to stay as string type

for col in cols:
    hfa[col] = hfa[col].astype("str")
    hfa[col] = hfa[col].str.strip()
    hfa[col] = pd.to_numeric(hfa[col], errors="ignore")
```

```python
#write funtion to replace strings:

def replace(column, old_str, new_str):
    hfa[column] = hfa[column].astype("str")
    hfa[column] = hfa[column].str.replace(old_str, new_str)
    hfa[column] = pd.to_numeric(hfa[column], errors="ignore")
```

```python
#loop through all columns and replace "---" with ""

for col in cols:
    replace(col,"---","")
```

```python
hfa.info()
#I did a manual check on the Health Facility Assessment Excel file using ISNUMBER()
#and there are indeed 71 columns with numerical data 
#so this is now showing correct data types
```

```python
#check for completely blank columns
list_of_null = hfa.isnull().sum().to_list()
list_of_null
```

```python
#drop empty column
hfa = hfa.dropna(axis="columns", how="all")
```

```python
#redefine list of columns after dropping one of them
cols = hfa.columns.to_list()
```

```python
#last col has 6 null values - replace them with 0
hfa["form.score_percent"] = hfa["form.score_percent"].astype("str")
hfa["form.score_percent"] = hfa["form.score_percent"].str.replace("nan", "0")
hfa["form.score_percent"] = pd.to_numeric(hfa["form.score_percent"], errors="ignore")
```

```python
#get list of which columns are numeric
num_col_list = []
for col in cols:
    is_num = is_numeric_dtype(hfa[col])
    if is_num == True:
        num_col_list.append(col)
```

```python
#check the list: length and first few columns
print(len(num_col_list))
print(num_col_list[0:5])
```

```python
#get list of which columns are non categorical
cat_col_list = []
for col in cols:
    is_cat = is_string_dtype(hfa[col])
    if is_cat == True:
        cat_col_list.append(col)
```

```python
#check the list worked: length and first few columns
print(len(cat_col_list))
print(cat_col_list[0:5])
```

```python
#encode categorical data using Ordinal Encoder
ordinal_encoder = OrdinalEncoder()
hfa[cat_col_list] = ordinal_encoder.fit_transform(hfa[cat_col_list])
```

```python
#set target
y = hfa["form.score_percent"]
```
